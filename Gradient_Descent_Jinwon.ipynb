```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**2

def df(x):
    return 2*x  
```

```python
def gradient_descent(start_x, lr=0.1, steps=25):
    x = start_x
    history = [x]

    for i in range(steps):
        grad = df(x)
        x = x - lr * grad
        history.append(x)
        
    return history

history = gradient_descent(start_x=6, lr=0.1, steps=25)
len(history)
```

```python
history = gradient_descent(start_x=6, lr=0.1, steps=20)
```

```python
xs = np.linspace(-6, 6, 300)
ys = f(xs)

plt.plot(xs, ys, label="f(x) = x^2")

plt.scatter(history[0], f(history[0]), c="red", s=200, label="Start")

for i, x in enumerate(history):
    plt.scatter(x, f(x), c="blue", s=60)
    plt.text(x, f(x)+0.5, str(i), fontsize=9)

plt.title("Gradient Descent Optimization Path")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```
